```python
import cx_Oracle
```

* cx_Oracle.makedsn : 오라클에 대한 주소 정보<br>
* cx_Oracle.connect : 오라클 접속 유저 정보

```python
#java에서 jsp에 해당, 1521은 코드 번호
dsn = cx_Oracle.makedsn('localhost', 1521, 'orcl')

#id, pw 적어서 연결
db = cx_Oracle.connect('scott', 'tiger')
```

* db.cursor : 데이터 담을 메모리의 이름을 선언<br>
* cursor.execute : SQL의 결과가 cursor 메모리를 담는다

```python
cursor = db.cursor()
cursor.execute("select * from emp")
```

* cursor.fetchall : 메모리에 담긴 데이터를 한 행씩 fetch 한다 (전부 all)

```python
data = cursor.fetchall()
data
```

```python
df = pd.DataFrame(data)
df
```

* cursor.description : 데이터의 칼럼명을 추출

```python
#칼럼명 추출
colnames = cursor.description

col = []

for colname in colnames:
    col.append(colname[0])
    
print(col)
print()

df.columns = col
df
```
