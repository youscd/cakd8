```python
meat_to_animal = {'bacon' : 'pig', 'pulled pork' : 'pig',
                  'pastrami' : 'cow', 'corned beef' : 'cow',
                 'honey ham' : 'pig', 'nova lox' : 'salmon'}
```

```python
#food 값을 먼저 소문자로 바꾸지 않고
#data['food'].map(lambda x: meat_to_animal[x.lower()])
data['food'].str.lower().map(meat_to_animal)
```

```python
data['animal'] = data['food'].str.lower().map(meat_to_animal)
data
```

```python
#replace 를 이용해서도 가능하다
```

```python
data = pd.DataFrame({'food' : ['bacon', 'pulled pork', 'bacon',
                               'Pastrami', 'corned beef', 'Bacon',
                               'pastrami', 'honey ham', 'nova lox'],
                     'ounces' : [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

#### (2) 두 데이터 프레임 합치기 merge

```python
df1 = pd.DataFrame({'key' : ['b','b','a','c','a','a','b'], 'data1' : range(7)})
df2 = pd.DataFrame({'key' : ['a','b','d'], 'data2' : range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how = 'inner')
pd.merge(df1, df2, how = 'outer')
```

```python
pd.merge(df1, df2, how = 'outer', indicator = True)
```

```python
pd.merge(df1, df2, on = 'key', how = 'outer', indicator = True)
```

```python
pd.merge(df1, df2, on = 'key', how = 'left', indicator = True)
```

```python
pd.merge(df1, df2, on = 'key', how = 'right', indicator = True)
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on = 'lkey')
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right)
```

```python
pd.merge(left, right, on = 'key1')
```

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b']) 
```
