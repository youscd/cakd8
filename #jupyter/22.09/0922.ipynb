- 수염의 길이 : (3사분위 + 1.5IQR) - (1사분위 - 1.5IQR)

```python
welfare = raw_welfare.loc[:, ['h14_g3','h14_g4','h14_g10','h14_g11','p1402_8aq1','h14_eco9','h14_reg7']]
welfare
```

```python
welfare = welfare.rename(columns = {'h14_g3':'gender', 'h14_g4':'birth', 'h14_g10':'marriage_type',
              'h14_g11':'religion', 'p1402_8aq1':'income',
              'h14_eco9':'code_job', 'h14_reg7':'code_region'})
welfare
```

```python
welfare['gender'].value_counts()
welfare['gender'].unique()
```

```python
welfare['gender'] = np.where(welfare['gender'] == 1.0, 'male', 'female')
```

```python
welfare.groupby('gender')[['income']].mean()
```

```python
sns.barplot(data = df, x = 'gender', y = 'income')
```

#### (2) 나이와 월급의 관계 (나이별 월급 평균)

```python
welfare.head()
```

```python
print(min(welfare.birth))
print(max(welfare.birth))
```

```python
welfare['age'] = 2019.0 - welfare['birth']
welfare[:2]
```

```python
welfare.isna().sum()
```

```python
welfare['age'] = welfare['age'].astype('int')
welfare.info()
```

```python
df1 = welfare.groupby('age', as_index = False)[['income']].mean()
df1
```

```python
df1.isna().sum()
```

```python
sns.lineplot(data = df1, x = 'age', y = 'income')
```

#### (3) 연령대에 따른 월급의 차이<br>
칼럼 이름: age_gen / 30세 이하 : young / 60세 이하 : middle / 나머지 : old

```python
welfare['age_gen'] = np.where(welfare['age'] <= 30,\
                              'young', np.where(welfare['age'] <= 60,\
                                                'middle', 'old'))
welfare[:2]
```

```python
welfare['age_gen'] = welfare['age_gen'].astype('category')
welfare.info()
```

```python
#df_gen = df_gen.reset_index()
df_gen = welfare.groupby(['age_gen', 'gender'], as_index = False)[['income']].mean()
df_gen
```

```python
#세대별 소득평균
sns.barplot(data = df_gen, x = 'age_gen', y = 'income', order = ['young', 'middle', 'old'], palette = 'Set1')
```

```python
#세대별, 성별 소득평균
sns.barplot(data = df_gen, x = 'age_gen', y = 'income', order = ['young', 'middle', 'old'], hue = 'gender')
```

#### (1) 대통령 연설문 텍스트 마이닝

```python
with open("speech_moon.txt", 'r', encoding = 'UTF-8') as f:
    moon = f.read()
```

```python
moon = open("speech_moon.txt", 'r', encoding = 'UTF-8').read()
moon
```

```python
import re
p = re.compile("[^가-힣]")
p.sub("  ", moon)
```

```python
moon = p.sub("  ", moon)
moon
```

```python
import konlpy
```

```python
hannanum = konlpy.tag.Hannanum()
```

```python
nouns = hannanum.nouns(moon)
nouns
```

```python
df_word = pd.DataFrame(nouns)
df_word
```

```python
#글자수 세기 : str로 변환 후 len()으로 확인
df_word.iloc[:,0].str.len()
```

```python
df_word['count'] = df_word.iloc[:,0].str.len()
df_word[:2]
```

```python
df_word.rename({0:'word'}, axis = 1, inplace = True)
df_word[:4]
```

```python
df_word = df_word[df_word['count'] >= 2]
df_word
```

```python
#단어의 빈도 구하기
df_word.groupby('word', as_index = False).agg('count').sort_values('count', ascending = False)
```

```python
top20 = df_word.groupby('word', as_index = False).agg('count').sort_values('count', ascending = False).head(20)
top20
```

```python
sns.barplot(data = top20, x = 'word', y = 'count')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
sns.barplot(data = top20, x = 'word', y = 'count')
plt.xticks(rotation = 20)
```
