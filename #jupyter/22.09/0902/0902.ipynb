## 조건문 if
- 구조 :
```
if 조건 :
    참일 때 실행될 문장
elif:
    참일 때 실행될 문장
elif:
    참일 때 실행될 문장
else:
    나머지 모두에서 실행될 문장
```

````python
if 1 < 2 :
    print("yes")
````

````python
if 1 > 2 :
    print("yes")
else :
    print("no")
````

````python
#온도(temp)가 25도보다 크면 "여름", 15보다 크면 "가을", 5도보다 크면 "봄", 나머지는 "겨울"
temp = int(input("온도: "))
if temp > 25:
    print("여름")
elif temp > 15:
    print("가을")
elif temp > 5:
    print("봄")
else:
    print("겨울")
````

````python
#입력받은 숫자가 홀수인지, 짝수인지 판단
number = int(input("숫자 입력: "))
if number%2 == 0:
    print("짝수 입니다")
else:
    print("홀수 입니다")
````
