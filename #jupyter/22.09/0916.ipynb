```python
data = '''
안녕하세요 그리고 그리고
안녕하세요 또 또
안녕하세요 그런데 그런데 
안녕하세요 끝 끝
'''

p = re.compile(r"(?P<repeat>\w+)\s+(?P=repeat)")

for d in p.finditer(data):
    print(d.group())
```

```python
text1 = '''
http://www.google.com/

https://mail.google.com/

ftp://ftp.google.com/

''' 
```

```python
#['http:', 'https:', 'ftp:']
p = re.compile(".+:")
p.findall(text1)
```

```python
#['http', 'https', 'ftp']
p = re.compile(".+(?=:)")
p.findall(text1)
```

```python
text2 = '''apple: $50.24
banana: $35.25
pizza: $100
'''
```

```python
#['50.24', '35.25', '100']
p = re.compile("[$].+")
p.findall(text2)
```

```python
p = re.compile("\$.+")
p.findall(text2)
```

```python
p = re.compile("(?<=[$]).+")
p.findall(text2)
```

```python
#일반화_python
def add_list(n):
    li1 = list(range(1, n+1))
    li2 = list(range(1, n+1))
    
    li3 = []
    for i in range(n):
        li3.append(li1[i] + li2[i])
    
    return li3
```

```python
add_list(10)
```

```python
#일반화_numpy
def add_array(n):
    ar1 = np.array(range(1, n+1))
    ar2 = np.array(range(1, n+1))
    ar3 = ar1 + ar2

    return ar3
```

```python
add_array(10)
```

```python
def add_list():
    li1 = list(range(1, 101))
    li2 = list(range(1, 101))
    
    li3 = []
    for i in range(100):
        li3.append(li1[i] + li2[i])
    
    return li3
```
