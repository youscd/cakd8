### (1) 붓꽃 품종 예측하기

```python
from sklearn.datasets import load_iris
```

```python
load_iris()    #dataset의 형태
```

```python
iris = load_iris()
iris
```

```python
#key값은 몇 개인가?
iris.keys()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
iris
type(iris)    #type가 dict는 아니지만, dict와 같은 형태이므로 dict라고 봐도 무관하다
```

```python
iris.values()
iris.items()
```

```python
#접근 : key 값을 입력하면 그 key에 맞는 value값을 출력해준다
iris['data']
```

```python
iris_data = iris['data']
iris_label = iris['target']    #label : '이름' 이라는 의미
```

```python
train_test_split(iris_data, iris_label)    #train_test_split() : 두 조각으로(75%, 25%) 나눈다
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label)    #총 4조각이 된다
```

```python
y_test
```

```python
dt_clf = DecisionTreeClassifier()    #DecisionTreeClassifier(): 대문자로 시작 -> class
```

```python
dt_clf.fit(X_train, y_train)    #.fit() : X_train과 y_train의 패턴을 분석하고, dt_clf에 저장한다
#fitting : 패턴을 찾는다, 학습한다, 적합하다, 훈련하다
```

```python
X_train.shape    #.shape -> (row의 개수, column의 개수)
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
dt_clf.predict(X_test)
```

```python
y_test
```

```python
#X_test와 y_test 의 값 비교(예측)
y_pred = dt_clf.predict(X_test)
y_pred == y_test
```

```python
sum(y_pred == y_test)
```

```python
iris['target_names']
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)   #두 함수를 비교해서 몇개 맞았는지 확인 (정확도)
```
