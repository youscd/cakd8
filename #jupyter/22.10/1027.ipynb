```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

text = 'I decided, very early on, just to accept life unconditionally.I never expected it to do anything special for me, yet I seemed to accomplish far more than I had ever hoped. Most of the time it just happened to me withoutmy ever seeking it.'

sent_anal = SentimentIntensityAnalyzer()
sent_anal.polarity_scores(text)
```

```python
Surprise 에서의 예측 방법 - predict(), test()
test() : 데이터 세트 전체에 대해서 추천을 예측
predict() : 개별 사용자의 아이템에 대한 추천 평점 예측
```

```python
import pandas as pd
from surprise import Reader, Dataset, SVD, accuracy
from surprise.model_selection import train_test_split

ratings = pd.read_csv('./ratings.csv')

reader = Reader(rating_scale = (0.5,5.0)) 

data = Dataset.load_from_df(ratings[['userId','movieId','rating']], reader)

trainset, testset = train_test_split(data, test_size=0.25)

model = SVD(n_epochs=50, n_factors=50)
```

```python
#평점 출력
model.fit(trainset)
model.test(testset)
```

```python
#userID = 196 이 예측한  movieID=302 의 예측 평점
model.fit(trainset)
print(model.predict("196", "302"))
print(model.predict("196", "302")[3])
```

### ③ 출력 함수 생성

```python
def search_news_1(keyword, page=10):
    main_url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query='
    keyword = keyword
    sub_url = '&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=31&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all'
    page_number = '&p='

    for i in range(1, page):
        url = main_url + keyword + sub_url + page_number + str(i)
        html = requests.get(url).text
        soup = BeautifulSoup(html, 'html.parser')
        news_tit = soup.find_all(class_ = "news_tit")

        for j in news_tit:
            print(j.text, ':', j.attrs['href'])
        print(f"{i}page 입니다")
        print("-" * 100)
        print()
```

```python
def search_news_2(keyword, page=10):
    main_url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query='
    keyword = keyword
    sub_url = '&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=31&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all'
    page_number = '&p='
    
    lst_1 = []
    lst_2 = []

    for i in range(1, page):
        url = main_url + keyword + sub_url + page_number + str(i)
        html = requests.get(url).text
        soup = BeautifulSoup(html, 'html.parser')
        news_tit = soup.find_all(class_ = "news_tit")

        for j in news_tit:
            lst_1.append(j.text)
            lst_2.append(j.attrs['href'])
        #print(lst_1)
        #print(lst_2)
        data = pd.DataFrame({"제목" : lst_1, "링크" : lst_2})
        
    return data
```

#### * csv 파일로 저장하기

```python
def search_news_3(keyword, page=10):
    keyword = keyword
    main_url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query='
    sub_url = '&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=31&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all'
    page_number = '&p='
    
    lst_1 = []
    lst_2 = []

    for i in range(0, page):
        url = main_url + keyword + sub_url + page_number + str(i)
        html = requests.get(url).text
        soup = BeautifulSoup(html, 'html.parser')
        news_tit = soup.find_all(class_ = "news_tit")

        for j in news_tit:
            lst_1.append(j.text)
            lst_2.append(j.attrs['href'])
        #print(lst_1)
        #print(lst_2)
        data = pd.DataFrame({"제목" : lst_1, "링크" : lst_2})
        data.to_csv(f"{keyword}.csv")
        
    return data
```

```python
driver = webdriver.Chrome("C:/Users/admin/chromedriver_win32/chromedriver.exe")

url1 = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query='
url2 = quote_plus("블랙핑크")    #quote_plus( ) : 한글을 처리해주는 함수
url = url1 + url2

driver.get(url)
time.sleep(5)     #무조건 5초는 기다리기
#tiem.implicitly_wait(10)    #최대 10초까지 기다리기

html = driver.page_source
driver.close()    #driver 끄기

soup = BeautifulSoup(html, 'html.parser')
images = soup.find_all(class_ = '_image _listImage')

count = 0
for image in images:
    urlretrieve(image['src'], f"./#images/{count}.jpg")
    count += 1
```
