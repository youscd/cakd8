```python
import pandas as pd
import openpyxl

GwanGwang = pd.read_excel('jeju/jejudoGwanGwang.xlsx')
jejudoMatJip = pd.read_excel('jeju/jejudoMatJip.xlsx')
jejuMatJip = pd.read_excel('jeju\jejuMatJip.xlsx')
jejuYeoHang = pd.read_excel('jeju\jejuYeoHang.xlsx')
```

```python
# 중복데이터는 거르고 결합
# df1.combine_first(df2) ====> df1을 기준으로 df2를 중복없이 결합하겠다.
jejuGwanGwang1 = GwanGwang.combine_first(jejudoMatJip)
jejuGwanGwang1 = jejuGwanGwang1.combine_first(jejuMatJip)
jejuGwanGwang1 = jejuGwanGwang1.combine_first(jejuYeoHang)
jejuGwanGwang1
```

```python
# API 테스트
import requests
url = 'http://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format('합정 스타벅스')
headers = {
    "Authorization" : "KakaoAK 095562b7356dfd5360e2c6860f69d4b2"
}
palces = requests.get(url, headers=headers).json()['documents']
palces[0]
# y 위도
# x 경도
# address_name : 위치
```

```python
def place_find(s) :
    # 접속 url
    url = 'http://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(s)
    
    # headers 입력
    headers = {
        # 카카오지도 접속
        "Authorization" : "KakaoAK 095562b7356dfd5360e2c6860f69d4b2"
    }
    # API 요청
    palces = requests.get(url, headers=headers).json()['documents']
    try :
        palces = palces[0]
        jejuGwanGwang1_name  = palces['place_name']
        위도 = palces['y']
        경도 = palces['x']
        
        result = [jejuGwanGwang1_name, 위도, 경도]
        
        return result
    except:
        pass
```

```python
place_find('청초수물회&섭국 [신관]')
```

```python
# 결측치 제거
a = jejuGwanGwang1['place'].dropna()
a
```

```python
a.value_counts()
```

```python
jeju_palce_list = []
for a in a :
    try :
        data = place_find(a)
        jeju_palce_list.append(data)
    except :
        pass
```

```python
jeju_palce_list
```

```python
jeju_palce_list_1 = jeju_palce_list
```

```python
#결측치 제거
jeju_palce_list_1 =  [x for x in jeju_palce_list_1 if x is not None]
jeju_palce_list_1
```

```python
jeju_palce = pd.DataFrame(jeju_palce_list_1)
jeju_palce
```

```python
jeju_palce.columns = ['place_name', '위도', '경도']
```

```python
jeju_palce = jeju_palce[(jeju_palce['place_name'] != '제주도')]
jeju_palce
```

```python
jeju_palce.drop_duplicates()
jeju_palce = jeju_palce.drop_duplicates()
```

```python
import folium
jejuMap = folium.Map(location=[33.36, 126.52], zoom_start = 10)
```

```python
for i in range(len(jeju_palce)) :
    # print(jeju_palce.iloc[i]['위도'])
    # print(jeju_palce.iloc[i]['경도'])
    folium.Marker([jeju_palce.iloc[i]['위도'], jeju_palce.iloc[i]['경도']],
                  popup = jeju_palce.iloc[i]['place_name'],
                  icon = folium.Icon(color = 'blue')).add_to(jejuMap)
jejuMap
```

```python
import folium

from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(jejuMap)

for i in range(len(jeju_palce)):
    folium.Marker(
        location = [jeju_palce.iloc[i]['위도'], jeju_palce.iloc[i]['경도']],
        popup = jeju_palce.iloc[i]['place_name'],
        icon = folium.Icon(color='blue',icon='ok'),
    ).add_to(marker_cluster)

jejuMap.save('jeju.html')
```
