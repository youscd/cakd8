## 9. 추천 시스템
추천 시스템의 대표적인 방식으로는 '콘텐츠 기반 필터링' 과 '협업 필터링' 이 있다

#### 콘텐츠 기반 필터링
아이템(상품, 영화, 서비스 등)을 구성하는 여러 가지 콘텐츠 중 사용자가 좋아하는 콘텐츠를 필터링하여 이에 맞는 아이템을 추천하는 방식이다<br>
예: 사용자가 좋아하는 영화를 분석한 뒤, 그 영화의 콘텐츠(장르, 감독, 영화배우 등)와 유사한 다른 영화를 추천하는 것<br>
콘텐츠 요소들을 결합하여 하나의 콘텐츠 특징으로 피처 벡터화한 후에 이 피처 벡터와 가장 유사한 피처 벡터를 가진 다른 영화를 추천하는 것<br>

#### 협업 필터링 - 최근접 이웃 협업 필터링, 잠재 요인 협업 필터링
* 최근접 이웃 협업 필터링<br>
사용자 기반과 아이템 기반 방식으로 나뉜다<br>
아이템 기반 최근접 이웃 방식은 특정 아이템과 가장 유사한 다른 아이템을 추천하는 것이다<br>
이때 유사도의 기준이 되는 것은 사용자들의 아이템에 대한 평가를 벡터화한 값이다 (코사인 유사도를 이용한다)<br>

* 잠재 요인 협업 필터링<br>
사용자-아이템 평점 행렬 데이터에 숨어있는 잠재 요인을 추출하여 사용자가 아직 평점을 매기지 않은 아이템에 대한 평점을 예측하고 이를 추천에 반영하는 방식이다<br>
다차원의 사용자-아이템 평점 행렬을 저차원의 사용자-잠재요인, 아이템-잠재요인 행렬로 분해하는 '행렬 분해' 기법을 사용한다

#### ③ 장르 콘텐츠 필터링을 이용한 영화 추천
장르 유사도에 따라 영화를 추천하는 함수 find_sim_movie() 생성

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    
    # 인자로 입력된 movies_df 에서 'title' 컬럼이 입력된 title_name 값인 DataFrame추출
    title_movie = df[df['title'] == title_name]
    
    # title_name을 가진 DataFrame의 index 객체를 ndarray로 반환하고 
    # sorted_ind 인자로 입력된 genre_sim_sorted_ind 객체에서 유사도 순으로 top_n 개의 index 추출
    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]
    
    # 추출된 top_n index들 출력. top_n index는 2차원 데이터 임. 
    #dataframe에서 index로 사용하기 위해서 1차원 array로 변경
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)
    
    return df.iloc[similar_indexes]
```

* 가중 평점<br>
(v / (v+m)) * R + (m / (v+m)) * C<br>

v : 개별 영화에 평점을 투표한 횟수<br>
m : 평점을 부여하기 위한 최소 투표 횟수<br>
R : 개별 영화에 대한 평균 평점<br>
C : 전체 영화에 대한 평균 평점

```python
#평점에 평가 횟수를 반영하는 새로운 평가 방식 -> 가중평점 사용
C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6)
print(C, m)
```

* 기존 평점을 새로운 가중 평점으로 변경하는 함수 weighted_vote_average() 생성<br>
새로운 평점 정보 : 'vote_weighted'

```python
def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ( (v / (v+m)) * R ) + ( (m / (v+m)) * C )
```

* 새롭게 정의된 평점 기준에 따라 다시 영화 추천

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    
    # top_n의 2배에 해당하는 쟝르 유사성이 높은 index 추출 
    similar_indexes = sorted_ind[title_index, :(top_n*2)]
    similar_indexes = similar_indexes.reshape(-1)
    
    # 기준 영화 index는 제외
    similar_indexes = similar_indexes[similar_indexes != title_index]
    
    # top_n의 2배에 해당하는 후보군에서 weighted_vote 높은 순으로 top_n 만큼 추출 
    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending=False)[:top_n]

similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather',10)
similar_movies[['title', 'vote_average', 'weighted_vote']]
```

```python
from surprise import SVD
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import train_test_split
```

```python
data = Dataset.load_builtin('ml-100k')
```

```python
#학습, 데이터 세트로 분리
trainset, testset = train_test_split(data, test_size=.25, random_state=0)
```

```python
#SVD로 잠재 요인 협업 필터링 수행
algo = SVD(random_state=0)
algo.fit(trainset)
```

```python
#추천을 예측하는 메서드1 : test() 사용자-아이템 평점 데이터 세트 전체에 대해서 추천을 예측
predictions = algo.test(testset)

print('prediction type:', type(predictions), 'size:', len(predictions))
predictions[:5]
```

```python
#uid : 사용자 아이디, iid : 아이템 아이디, est : 추천 예측 평점 (r_ui : 실제 평점)
[ (pred.uid, pred.iid, pred.est) for pred in predictions ]
[ (pred.uid, pred.iid, pred.est) for pred in predictions[:3] ]
```

```python
#추천을 예측하는 메서드2 : predict() 개별 사용자의 아이템에 대한 추천 평점 예측
#사용자 아이디, 아이템 아이디는 문자열로 입력
uid = str(196)
iid = str(302)
pred = algo.predict(uid, iid)
pred
```

```python
#테스트 데이터 세트를 통해 추천 예측 평점과 실제 평점과의 차이 평가
accuracy.rmse(predictions)
```

#### * Surprise 를 이용한 개인화 영화 추천 시스템 구축
Surprise 패키지로 학습된 추천 알고리즘을 기반으로 특정 사용자가 아직 평점을 매기지 않은 (관람하지 않은) 영화 중에서 개인 취향에 가장 적절한 영화를 추천해보자

데이터 세트 전체를 학습 데이터로 사용하기 위해 DatasetAutoFolds 클래스를 이용한다<br>
DatasetAutoFolds 객체의 build_full_trainset() 메서드를 호출하여 전체 데이터 세트를 학습 데이터로 사용한다

```python
from surprise.dataset import DatasetAutoFolds
```

```python
reader = Reader(line_format = 'user item rating timestamp', sep=',', rating_scale=(0.5, 5))

#DatasetAutoFolds 클래스를 ratings_noh.csv 파일 기반으로 생성
data_folds = DatasetAutoFolds(ratings_file='ratings_noh.csv', reader= reader)

#전체 데이터를 학습 데이터로 생성
trainset = data_folds.build_full_trainset()
```

```python
#SVD를 이용해 학습 수행
algo = SVD(n_epochs=20, n_factors=50, random_state=0)
algo.fit(trainset)
```

```python
#영화에 대한 상세 속성 정보 DataFrame 로딩
movies = pd.read_csv('./ml-latest-small/ml-latest-small/movies.csv')
movies
```

```python
#임의로 지정한 특정 사용자(userId=9), userId=9가 아직 평점을 매기지 않은 영화 movieId=24
movieIds = ratings[ratings['userId']==9]['movieId']
print(movieIds)

if movieIds[movieIds==42].count() == 0:
    print('userId=9 는 movieId=24 의 평점이 없음')
    
print(movies[movies['movieId'] == 42])
```

```python
#movieId=24 에 대한 userId=9 의 추천 예상 평점 확인 (est 값)
uid = "9"
iid = "42"
pred = algo.predict(uid, iid, verbose=True)
```

```python
#사용자가 평점을 매기지 않은 전체 영화에 대한 예측 평점
def get_unseen_surprise(ratings, movies, userId):
    #입력값으로 들어온 userId에 해당하는 사용자가 평점을 매긴 모든 영화 리스트
    seen_movies = ratings[ratings['userId'] == userId]['movieId'].tolist()
    
    #모든 영화의 movieId 리스트
    total_movieIds = movies['movieId'].tolist()
    
    #모든 영화의 movieId 중 이미 평점을 매긴 영화의 movieId는 제외
    #unseen_movies = [movie for movie in total_movieIds if movie not in seen_movies]
    unseen_movies = []
    for movie in total_movieIds:
        if movie not in seen_movies:
            unseen_movies.append(movie)
            
    print('전체 영화 수:', len(total_movieIds), '평점 매긴 영화 수:', len(seen_movies),
          '추천 대상 영화 수:', len(unseen_movies))
    
    return unseen_movies
```

```python
unseen_movies = get_unseen_surprise(ratings, movies, 9)
```

```python
#SVD를 이용해 높은 예측 평점을 가진 순으로 영화 추천
def recomm_movie_by_surprise(algo, userId, unseen_movies, top_n=10):
    
    #알고리즘 객체의 predict() 메서드를 평점이 없는 영화에 반복 수행
    predictions = [algo.predict(str(userId), str(movieId)) for movieId in unseen_movies]
    
    #predictions list 객체는 surprise의 Predictions 객체를 원소로 갖고 있다
    # [Prediction(uid='9', iid='1', est=3.69), Prediction(uid='9', iid='2', est=2.98),,,,]
    
    #predictions list 객체를 est 값으로 정렬
    def sortkey_est(pred):
        return pred.est
    
    predictions.sort(key=sortkey_est, reverse=True)
    top_predictions = predictions[:top_n]
    
    #top_n으로 추출된 영화의 정보 추출
    top_movie_ids = [int(pred.iid) for pred in top_predictions]
    top_movie_titles = movies[movies.movieId.isin(top_movie_ids)]['title']
    top_movie_rating = [pred.est for pred in top_predictions]
    
    top_movie_preds = [(id, title, rating) for id, title, rating in \
                      zip(top_movie_ids, top_movie_titles, top_movie_rating)]
    
    return top_movie_preds
```

```python
unseen_movies = get_unseen_surprise(ratings, movies, 9)
top_movie_preds = recomm_movie_by_surprise(algo, 9, unseen_movies, top_n=10)
print('##### Top-10 추전 영화 리스트 #####')

for top_movie in top_movie_preds:
    print(top_movie[1], ":", top_movie[2])
```
