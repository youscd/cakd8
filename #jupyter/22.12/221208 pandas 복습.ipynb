### (1) 데이터 프레임 생성

```python
## 데이터 프레임 만들기
pd.DataFrame(data = {'국어' : [80, 70], '영어' : [90, 80], '수학' : [100, 90]},
            index = ['A','B'])
```

```python
## 과학 칼럼 90, 90 추가
df = pd.DataFrame(data = {'국어' : [80, 70], '영어' : [90, 80], '수학' : [100, 90]},
                  index = ['A','B'])

df['과학'] = [90, 90]
df
```

```python
## 수학 칼럼 삭제
df.drop(columns = '수학', inplace=True)
df
```

```python
## 총점 칼럼 추가
df['총점'] = df['국어'] + df['영어'] + df['과학']
df
```

```python
## 행이름 C, 점수 70 80 90 240 추가
df.loc['C'] = [70, 80, 90, 240]
df
```

```python
## total의 260 을 결측치로
df.replace(df['총점'][0], np.nan, inplace=True)
```

```python
df.replace(df['총점'][0], 260, inplace=True)
df
```

```python
## 칼럼 이름 바꾸기
df.rename(columns = {'total', '총점'})
#다른방법 : df.columns = ['국어', '영어', '과학', '총점']
```

```python
len(df.columns)
```

```python
## 평균이 85 이상이면 A+, 이하이면 A-
df['평균'] = df['총점'] / (len(df.columns) - 1)
df['학점'] = np.where(df['평균'] >= 85, 'A+', 'A-')
df
```

```python
## 총점, 국어 의 오름차순으로 정렬
df.sort_values(by = ['총점', '국어'])
```

```python
## mpg['class'] 의 데이터 타입을 'category' 로 변경
mpg.info()
```

```python
mpg['class'] = mpg['class'].astype('category')
mpg.info()
```

```python
mpg['cty'][4]
```

```python
## mpg['cty'] 값이 짝수이면 2를 더하고, 홀수이면 1을 더한다
# 방법1
for i in range(len(mpg['cty'])):
    if mpg['cty'][i] % 2 == 0:
        mpg['cty'][i] += 2
    else:
        mpg['cty'][i] += 1
        
# 방법2
mpg['cty'] = np.where(mpg['cty'] % 2 == 1, mpg['cty'] + 1, mpg['cty'] + 2)
```

```python
mpg['cty']
```

```python
## map, filter, apply, ...
def add(x):
    if x % 2 == 0:
        return x+2
    else:
        return x+1

mpg['cty_1'] = list(map(add, mpg['cty']))
mpg[:5]
```

```python
list(filter(add, mpg['cty']))
```

```python
mpg['cty'].apply(lambda x : x+2 if x%2 == 0 else x+1)
```

```python
## mpg['hwy'] 값 분류 : ~20 - A, ~30 - B, ~40 - C, 40~ - D
# 방법1
mpg['hwy_rank'] = []

if mpg['hwy'] < 20:
    mpg['hwy_rank'].append('A')
elif mpg['hwy'] < 30:
    mpg['hwy_rank'].append('B')
elif mpg['hwy'] < 40:
    mpg['hwy_rank'].append('C')
else:
    mpg['hwy_rank'].append('D')
    
mpg[:5]
```

```python
# 방법2
mpg['hwy_rank'] = np.where(mpg['hwy'] <= 20, 'A',
                          (np.where(mpg['hwy'] <= 30, 'B',
                                   (np.where(mpg['hwy'] <= 40, 'C', 'D')))))
```

```python
# 방법3
mpg['hwy_rank'] = pd.cut(mpg['hwy'],
                         bins = [0, 20, 30, 40, 100],
                         labels=['A','B','C','D'])

mpg[:5]
```

```python
## mpg['trans'] 에서 'auto' 인 행 추출
mpg[mpg['trans'].str[:4] == 'auto']['trans'][:10]
```

```python
mpg[mpg['trans'].str.contains('auto')][:10]
```

```python
mpg = pydataset.data('mpg').reset_index().iloc[:,1:]
mpg['info'] = mpg['manufacturer'] + '/' + mpg['model'] + '/' + mpg['drv']
mpg[:5]
```

```python
## 칼럼 삭제
mpg.drop(columns = ['manufacturer', 'model', 'drv'], inplace=True)
mpg
```

```python
mpg['info'][3]
```

```python
mpg['info'].str.split('/')[1]
```

```python
## mpg['info'] 에서 'manufacturer', 'model', 'drv' 칼럼 생성

info = []
for i in range(len(mpg)):
    info.append(mpg['info'][i])
    
info[:5]
```

```python
info[0].split('/')[0]
info[0].split('/')[1]
```

```python
len(info)
```

```python
manufacturer = []
model = []
drv = []

for i in range(len(info)):
    #print(info[i].split('/')[0])
    manufacturer.append(info[i].split('/')[0])
    model.append(info[i].split('/')[1])
    drv.append(info[i].split('/')[2])

mpg['manufacturer'] = manufacturer
mpg['model'] = model
mpg['drv'] = drv

mpg[:10]
```
